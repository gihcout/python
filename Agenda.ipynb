Agenda de contatos contendo nome e celular de n pessoas, armazenados em um dicionário. Os dados devem ser lidos pelo teclado.


```python
dados = dict()

while True:
  try:
    print('Digite o n° da ação que deseja realizar\n')
    i = int(input('1- Inserir novo contato\n2- Alterar contato\n3- Deletar contato\n4- Consultar agenda\n5-Fechar agenda\n'))
    if i == 1:
      nome = input('Nome: ')
      dados[f'{nome}'] = input('Celular: ')
      print('\nContato registrado!\n')
    elif i == 2:
      op = int(input('Digite 1 para alterar o nome e 2 para alterar o número\n'))
      if op == 1:
        n = input(f'Informe qual contato deseja alterar o nome\n{dados.keys()}\n')
        nn = input('Digite o novo nome: ')
        dados[f'{nn}'] = dados.pop(f'{n}')
        print('Nome alterado!\n')
      else:
        c = input(f'Informe qual contato deseja alterar o numero\n{dados.keys()}\n')
        dados[f'{c}'] = input('Digite o novo numero: ')
        print('\nNúmero alterado!\n')
    elif i == 3:
      d = input(f'Informe qual contato deseja deletar\n{dados.keys()}\n')
      dados.pop(f'{d}')
      print('Contato deletado!\n')
    elif i == 4:
      for x,y in dados.items():
        print(f'Nome: {x}\nCelular: {y}')
    else:
      print('\nAgenda fechada\n')
      break
  except ValueError:
    print('Valor inválido.\n')
```
