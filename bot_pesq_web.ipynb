```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd 
import time
import re
```

```python
base = pd.read_excel('consulta.xlsx')
infos = []
```

```python
# Quando finalizado habilitar esse modo, para execução em modo headless
# Inicialize o driver passando as opções do Chrome
# chrome_options = Options()
# chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)

i = 0

while i < len(base):
    # Navegando até a página
    driver.get("https://meu.inss.gov.br/v35/index.html#/pericia")

    # Preenchendo os campos de acordo com o xlsx
    linha = base.iloc[i]

    # Input name
    text_area = driver.find_element("xpath",'/html/body/div[1]/div/div/div/div[3]/section/section/div/div[1]/div[2]/div[1]/div/span/input')
    text_area.send_keys(linha['nome'])

    # Input cpf
    text_area = driver.find_element("xpath",'/html/body/div[1]/div/div/div/div[3]/section/section/div/div[1]/div[2]/div[2]/div[1]/div/div/span/input')
    text_area.send_keys(str(linha['cpf']))

    # Input birthday
    text_area = driver.find_element("xpath",'/html/body/div[1]/div/div/div/div[3]/section/section/div/div[1]/div[2]/div[2]/div[2]/div/div/span/input')
    text_area.send_keys(linha['data nascimento'].strftime('%d/%m/%Y'))

    # Click "continuar"
    click_continue = driver.find_element("xpath",'/html/body/div[1]/div/div/div/div[3]/section/section/div/div[3]/div/button')
    click_continue.click()

    # try:
    #     # Para extrair o pdf do requerimento
    #     link_pdf = WebDriverWait(driver, 10).until(
    #     EC.presence_of_element_located((By.XPATH, '//i[@class="ico-doc-pdf" and @title="Gerar Comprovante"]'))
    #     )
    #     link_pdf.click()

    #     time.sleep(10)

    try:
        # Click da lupa, abrirá a nova janela com as informações que desejamos
        link_pdf = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//i[@class="ico-magnifier" and @title="Detalhar Requerimento"]'))
        )
        link_pdf.click()


        # Localiza os elementos que contêm as informações para extrair
        data_hora = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div/div/div/div[3]/section/section[1]/section/div[1]/div[1]/div[3]/span'))
        )
        endereco = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[3]/section/section[1]/section/div[1]/div[1]/div[5]/span')
        nb = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[3]/section/section[1]/section/div[1]/div[3]/div[5]/span')
        nr = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[3]/section/section[1]/section/div[1]/div[3]/div[6]/span')
        situacao = driver.find_element(By.XPATH, '/html/body/div[1]/div/div/div/div[3]/section/section[1]/section/div[1]/div[1]/div[2]/span')

        # # Extraia o texto
        info1 = data_hora.text
        info2 = endereco.text
        info3 = nb.text
        info4 = nr.text
        info5 = situacao.text

        # Organizando as informações
        infos_tratada = {
            'Nome': linha['nome'],
            'CPF': linha['cpf'],
            'Data de Nascimento': linha['data nascimento'],
            'data_hora': info1,
            'endereco': info2,
            'nb': info3,
            'nr': info4,
            'situacao': info5,
            'obs': None
        }
        infos.append(infos_tratada)

    except NoSuchElementException:
        pass

    # except NoSuchElementException:
    #     pass

    i += 1

# Fechando o driver após fazer todas as pesquisas
driver.quit()

# Converte a infos em um df
df = pd.DataFrame(infos)

# Salve o DataFrame em um novo arquivo Excel
df.to_excel('agendamentos_encontrados.xlsx', index=False)
```
